```python
"""
Simple Reflex Agent

Created Sept, 2025

@author: I_Osunmakinde
"""

import sqlite3
import random
import time


class SimpleReflexAgent:
    COLD_THRESHOLD = 20
    HOT_THRESHOLD = 25
    
    def environment(self):
        # generate the temp randomly and insert temporarily in a database                 
        self.temp_Data_Generator()                            
        
        rows = self.rows
        
        # Delay the process for seconds due to lag time
        delay_seconds = 2
        
        # perceive the environment temperature and decide
        for row in rows:
            self.perception_decision(row[1])            
            time.sleep(delay_seconds)
         
               
    def temp_Data_Generator(self):
        #Temperature Data Generation for the Environment.
        conn = sqlite3.connect("sragent")
        
        cur = conn.cursor()
        cur.execute('''CREATE TABLE IF NOT EXISTS temp_db
                        (temp_id INTEGER, tempCol REAL)''')       
                
        #Empty table
        cur.execute("DELETE FROM temp_db")
        conn.commit()
        
        #Generate random temperature values and insert                 
        for i in range(20):
            temp_value = random.uniform(0.0, 100.0)
            cur.execute('''INSERT INTO temp_db (temp_id, tempCol) VALUES (?, ?)''', (i, temp_value))
            conn.commit()               
        
        
        #Quick display
        cur.execute("SELECT * FROM temp_db")

        # Fetch all rows from the result set
        rows = cur.fetchall()
        self.rows = rows        
    
        cur.close()
        conn.close()
        
          
    def perception_decision(self, temp):        
        # Agent perceives and takes action based on temperature
        
        print(f"\n Current temperature = {temp:.2f}")

        if temp < self.COLD_THRESHOLD:
            print(" Cold Temperature → Turn on HEATER to warm the room.")
        elif temp > self.HOT_THRESHOLD:
            print(" Hot Temperature → Turn on AIR CONDITIONER to cool the room.")
        else:
            print(" Comfortable Temperature → Do nothing.")
            
        
        
        
    
    
    
    
    

#----------------------------Code Driver --------------------------------------    
agent = SimpleReflexAgent()

# Agent designed as semi-autonomous 
print("SIMPLE REFLEX AGENT ACTING ON ENVIRONMENT TEMPERATURE")
while True:    
    agent.environment()
    
    n = input("\n Press Enter-key to continue or enter 'q' to Quit: ")
    if n.strip().lower() == 'q':
        print("Please be safe in the environment...")
        break
    
    

```
